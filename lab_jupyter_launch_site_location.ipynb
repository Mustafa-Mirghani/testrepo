```python
# Task 2: Mark the success/failed launches for each site on the map
# Initialize the map at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add circles for launch sites as done earlier
for index, row in launch_sites_df.iterrows():
    coordinate = [row['Lat'], row['Long']]  # Get coordinates of each launch site
    site_name = row['Launch Site']  # Get the launch site name

    # Add circle for each launch site
    folium.Circle(
        coordinate,
        radius=1000,  # Radius of the circle
        color='#000000',  # Circle border color (black)
        fill=True,
        fill_color='#FFFF00',  # Fill color (yellow)
    ).add_child(folium.Popup(f'Launch Site: {site_name}')).add_to(site_map)

# Mark success/failed launches based on the 'class' column
for index, row in spacex_df.iterrows():
    launch_status = row['class']  # Get launch success/failure (1 for success, 0 for failure)
    coordinate = [row['Lat'], row['Long']]  # Get coordinates of the launch
    if launch_status == 1:
        # Success - Green marker
        folium.Marker(
            coordinate,
            icon=folium.Icon(color='green', icon='check'),
            popup=f"Launch Status: Success\n{row['Launch Site']}"
        ).add_to(site_map)
    else:
        # Failure - Red marker
        folium.Marker(
            coordinate,
            icon=folium.Icon(color='red', icon='times'),
            popup=f"Launch Status: Failure\n{row['Launch Site']}"
        ).add_to(site_map)

# Display the map with success/failure markers
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
